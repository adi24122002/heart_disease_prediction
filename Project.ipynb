Now we have got our data split into train and test sets, its time to build machine learning model

We'll train it (find the parameters) on the training set

and we'll test it (use the pattern) on the test set

We're going to try 3 different models
  1.Logistic Regression

  2.K Nearest Neighbour Classifier
  
  3.RandomForestClassifier
  

```python
#Put models in a dictionary
models = {"Logistic Regression": LogisticRegression(),
          "KNearest NEighbour": KNeighborsClassifier(),
          "RandomForestClassifier": RandomForestClassifier(n_estimators=100)}

#Create a function to fit and score the models
def fit_and_score(models,X_train, X_test, y_train, y_test):
  """
  Fits and evaluates given machine learning models
  models: a dict of different machine learning models
  X_train: training data (no labels)
  X_test: testing data (no labels)
  y_train: training labels
  y_test: test labels
  """

  #Set random seed
  np.random.seed(42)

  #Make a list to keep model scores
  model_scores = {}

  #Loop through models
  for name, model in models.items():
    #Fit the model to the data
    model.fit(X_train,y_train)
    #Evaluate the model and append its score to model_scores
    model_scores[name]=model.score(X_test,y_test)
  return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```

```python
#Model Comparison
model_compare = pd.DataFrame(model_scores, index=["Accuracy"])
model_compare.T.plot.bar();
```

#Hyperparameter Tuning

```python
#Lets tune KNN
train_scores = []
test_scores = []

#Create a list of different values for n_neighbours
neighbors = range(1,21)

#Setup KNN instance
knn =KNeighborsClassifier()

#Loop through different n_neighbours
for i in neighbors:
  knn.set_params(n_neighbors=i)

  #Fit the algo
  knn.fit(X_train,y_train)

  #Update the training score list
  train_scores.append(knn.score(X_train,y_train))

  #Update the test score list
  test_scores.append(knn.score(X_test,y_test))
```

```python
train_scores
```

```python
test_scores
```

```python
plt.plot(neighbors,train_scores,label="Train Score")
plt.plot(neighbors,test_scores,label="Test Score")
plt.xlabel("Number of Neighbors")
plt.ylabel("Model Score")
plt.legend()

print(f"Maximum KNN score on the test data: {max(test_scores)*100:.2f}%")
```

#Hyperparameter tuning with RandomizedSearchCV

* We're going to tune:
  1. RamdomFOrestClassifier

  2. Logistic Regression

```python
#Create a hyperparameter grid for Logistic Regression
log_reg_grid = {"C": np.logspace(-4,4,20),
                "solver":["liblinear"]}

#Create a hyperparameter grid for RandomForestClassification
rf_grid = {"n_estimators": np.arange(10,1000,500),
           "max_depth": [None,3,5,10],
           "min_samples_split": np.arange(2,20,2),
           "min_samples_leaf": np.arange(1,20,2)}
```

```python
np.logspace(-4,4,20)
```

Now we have got hyperparamter grids setup for each of our models,lets tune them usimg RandomizedSearchCV

```python
#Tune Logistic Regression

np.random.seed(42)

#Setup random hyperparameter search for LogisticRegreession
rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                                param_distributions= log_reg_grid,
                                cv=5,\
                                n_iter=20,
                                verbose=True)

#Fit random hyperparameter search for LogisticRegreession
rs_log_reg.fit(X_train,y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(X_test,y_test)
```

Now we've tuned Logistic Regression now we will do the smae for RandomForestClassifier

```python
#Setup random seed
np.random.seed(42)

#Setup random hyperparameter search for RandomForestClassifeir
rs_rf = RandomizedSearchCV(RandomForestClassifier(),
                           param_distributions= rf_grid,
                           cv=5,
                           n_iter=20,
                           verbose=True)

#Fit random hyperparameter search for RandomForestClassifier
rs_rf.fit(X_train,y_train)
```

```python
#Find the best parameters
rs_rf.best_params_
```

```python
#Evaluate the randomized search for randomforestclassifier
rs_rf.score(X_test,y_test)
```
